```python
from PIL import Image
```

```python
#put the picture in the content/..
img = Image.open("/content/getty.png")
img
```

```python
import google.generativeai as genai
from google.colab import userdata
genai.configure(api_key=userdata.get('API_KEY'))
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
```

```python
response = model.generate_content(img) #just sending image as prompt
```

```python
print(response.text)
```

```python
prompt= 'describe this image for me'
response = model.generate_content([img, prompt])
print(response.text)
```

```python
prompt= 'who is feeding the puppy'
response = model.generate_content([img, prompt])
print(response.text)
```
